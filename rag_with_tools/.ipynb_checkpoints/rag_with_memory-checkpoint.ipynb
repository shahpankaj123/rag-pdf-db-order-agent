```python
from langchain_groq import ChatGroq
from langchain.tools import tool
```

```python
import getpass
import os

if "groq_open_api_key" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")
```

```python
products = {
    "wireless headphones": {
        "price": "199",
        "description": "Best Bluetooth headphones with noise cancellation and long battery life"
    },
    "smartphone": {
        "price": "799",
        "description": "Latest 5G smartphone with high-resolution camera and fast processor"
    },
    "laptop": {
        "price": "1299",
        "description": "Powerful laptop suitable for programming, gaming, and productivity"
    },
    "smartwatch": {
        "price": "249",
        "description": "Fitness tracking smartwatch with heart rate monitoring and GPS"
    },
    "wireless mouse": {
        "price": "49",
        "description": "Ergonomic wireless mouse with adjustable DPI settings"
    },
    "mechanical keyboard": {
        "price": "129",
        "description": "RGB mechanical keyboard with tactile switches for gaming and typing"
    },
    "tablet": {
        "price": "499",
        "description": "Lightweight tablet with a large display and long battery life"
    },
    "bluetooth speaker": {
        "price": "99",
        "description": "Portable speaker with deep bass and waterproof design"
    },
    "gaming monitor": {
        "price": "349",
        "description": "27-inch monitor with 144Hz refresh rate and low response time"
    },
    "external SSD": {
        "price": "159",
        "description": "High-speed portable SSD with 1TB storage capacity"
    },
    "webcam": {
        "price": "79",
        "description": "Full HD webcam ideal for online meetings and streaming"
    },
    "printer": {
        "price": "199",
        "description": "All-in-one wireless printer with scanning and copying features"
    }
}

product_reviews = {
    "wireless headphones": {
        "reviews": 1250,
        "avg_rating": 4.7
    },
    "smartphone": {
        "reviews": 3420,
        "avg_rating": 4.6
    },
    "laptop": {
        "reviews": 1890,
        "avg_rating": 4.8
    },
    "smartwatch": {
        "reviews": 980,
        "avg_rating": 4.5
    },
    "wireless mouse": {
        "reviews": 2150,
        "avg_rating": 4.4
    },
    "mechanical keyboard": {
        "reviews": 1430,
        "avg_rating": 4.7
    },
    "tablet": {
        "reviews": 870,
        "avg_rating": 4.5
    },
    "bluetooth speaker": {
        "reviews": 1680,
        "avg_rating": 4.6
    },
    "gaming monitor": {
        "reviews": 1120,
        "avg_rating": 4.8
    },
    "external SSD": {
        "reviews": 760,
        "avg_rating": 4.9
    },
    "webcam": {
        "reviews": 540,
        "avg_rating": 4.3
    },
    "printer": {
        "reviews": 690,
        "avg_rating": 4.2
    }
}
```

```python
@tool
def get_product(name : str) -> str:
    """Look up a product by name and return its price and desciption"""
    p = products.get(name.lower())
    if p is None:
        return "Product Not Found"
    return str(p)

@tool
def get_product_rating_and_review(name : str) -> str:
    """Look up a product review by product name and return its number of review and avg rating"""
    p = product_reviews.get(name.lower())
    if p is None:
        return "Data Not Found"
    return str(p)
```

```python
from langchain.agents import create_agent

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0.8,
)

agent = create_agent(
    llm,
    tools = [get_product ,get_product_rating_and_review],
    system_prompt="You are a helpful products Assistant for an Online Tech Store",
)
```

```python
def ask(question : str):
    res = agent.invoke({"messages": [{"role": "user", "content": question}]})
    print(res['messages'][-1].content)
```

```python
ask("what is the price of smartphone")
```

```python
ask("what is the review of smartphone?")
```

```python
ask("Give me Description of last product?")
```

```python
from langgraph.checkpoint.memory import InMemorySaver  
```

```python
memory_agent = create_agent(
    llm,
    tools = [get_product ,get_product_rating_and_review],
    system_prompt="You are a helpful products Assistant for an Online Tech Store",
    checkpointer=InMemorySaver()
)
```

```python
def ask1(question : str):
    thread_config = {"configurable": {"thread_id": "1"}}
    res = memory_agent.invoke({"messages": [{"role": "user", "content": question}]},thread_config)
    print(res['messages'][-1].content)
```

```python
ask1("what is the price of smartphone")
```

```python
ask1("what are the review of this product?")
```

```python
ask1("which product do you know?")
```

```python
ask1("my name is pankaj shah")
```

```python
ask1("what's my name?")
```
